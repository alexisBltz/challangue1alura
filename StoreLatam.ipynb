### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
# Calcular ingresos por tienda
ingreso_tienda1 = tienda["Precio"].sum()
ingreso_tienda2 = tienda2["Precio"].sum()
ingreso_tienda3 = tienda3["Precio"].sum()
ingreso_tienda4 = tienda4["Precio"].sum()

# Mostrar resultados
print(f"Ingreso total Tienda 1: {ingreso_tienda1:,.2f}")
print(f"Ingreso total Tienda 2: {ingreso_tienda2:,.2f}")
print(f"Ingreso total Tienda 3: {ingreso_tienda3:,.2f}")
print(f"Ingreso total Tienda 4: {ingreso_tienda4:,.2f}")
```

# 2. Ventas por categoría

```python
# Función para contar productos por categoría en una tienda
def contar_por_categoria(df, nombre_tienda):
    conteo = df.groupby("Categoría del Producto").size().sort_values(ascending=False)
    print(f"\nCategorías más populares en {nombre_tienda}:")
    print(conteo)

# Aplicar a cada tienda
contar_por_categoria(tienda, "Tienda 1")
contar_por_categoria(tienda2, "Tienda 2")
contar_por_categoria(tienda3, "Tienda 3")
contar_por_categoria(tienda4, "Tienda 4")
```

# 3. Calificación promedio de la tienda


```python
# Calcular calificación promedio de cada tienda
promedio_tienda1 = tienda["Calificación"].mean()
promedio_tienda2 = tienda2["Calificación"].mean()
promedio_tienda3 = tienda3["Calificación"].mean()
promedio_tienda4 = tienda4["Calificación"].mean()

# Mostrar resultados con 2 decimales
print(f"Calificación promedio Tienda 1: {promedio_tienda1:.2f}")
print(f"Calificación promedio Tienda 2: {promedio_tienda2:.2f}")
print(f"Calificación promedio Tienda 3: {promedio_tienda3:.2f}")
print(f"Calificación promedio Tienda 4: {promedio_tienda4:.2f}")
```

# 4. Productos más y menos vendidos

```python
import matplotlib.pyplot as plt

# Función para mostrar top y bottom productos de una tienda
def productos_destacados(df, nombre_tienda):
    conteo = df.groupby("Producto").size().sort_values(ascending=False)

    # Top 5 más vendidos
    top5 = conteo.head(5)
    # Bottom 5 menos vendidos
    bottom5 = conteo.tail(5)

    print(f"\nProductos más vendidos en {nombre_tienda}:")
    print(top5)
    print(f"\nProductos menos vendidos en {nombre_tienda}:")
    print(bottom5)

    # Visualización
    fig, axes = plt.subplots(1, 2, figsize=(14,5))

    top5.plot(kind='bar', ax=axes[0], color='skyblue')
    axes[0].set_title(f"{nombre_tienda} - Top 5 más vendidos")
    axes[0].set_ylabel("Cantidad de ventas")
    axes[0].set_xlabel("Producto")

    bottom5.plot(kind='bar', ax=axes[1], color='salmon')
    axes[1].set_title(f"{nombre_tienda} - Top 5 menos vendidos")
    axes[1].set_ylabel("Cantidad de ventas")
    axes[1].set_xlabel("Producto")

    plt.tight_layout()
    plt.show()

# Aplicar a cada tienda
productos_destacados(tienda, "Tienda 1")
productos_destacados(tienda2, "Tienda 2")
productos_destacados(tienda3, "Tienda 3")
productos_destacados(tienda4, "Tienda 4")
```
